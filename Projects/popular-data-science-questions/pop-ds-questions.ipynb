## Most Used and Most Viewed

We are now focused on determining the most popular tags. To do this, we'll consider two different proxies for popularity. The first proxy is the number of times a tag was used. The second proxy is the number of times a question with that tag was viewed.

In order to further refine our research, we could take into account other factors such as the score of the question or whether it is a part of someone's favorite questions. However, for now, we'll limit our focus to counts and views as our proxies for popularity.

### Getting Tag count

Here, the goal is to count how often each tag was used. The following steps will be taken:

1. We start by creating an empty dictionary called `tag_count`.
2. Then, we loop through each *list of tags* in the `Tags` column of the `questions` dataframe.
3. For each *list of tags*, we loop through each tag within the list.
4. For each tag, we check if it already exists in the `tag_count` dictionary.
5. If the tag is already in the dictionary, we add 1 to its count.
6. If the tag isn't in the dictionary yet, we add it and start its count at 1.

````python
# Create an empty dictionary (tag_count) to store each tag as a key with its count as its value
tag_count = {}

for tags in questions['Tags']:
    for tag in tags:
        if tag in tag_count:
            tag_count[tag] += 1
        else:
            tag_count[tag] = 1
````

Let's turn `tag_count` into a dataframe (`most_used`) for a better display.

````python
# Create dataframe most_used
most_used = pd.DataFrame.from_dict(tag_count, orient='index', columns=['Count'])
most_used
````

Get the top 20 most used tags in the `most_used` dataframe.

````python
most_used = most_used.sort_values(by='Count', ascending=False).head(20)
most_used
````

Let's visualize the top 20 most used tags.

````python
ax = most_used.sort_values('Count').plot(kind='barh', figsize=(12, 6), color='darkblue')
ax.set_title('Top 20 Most Popular Tags on DSSE', size=20, weight='bold' )
ax.grid(False)

plt.show()
````

As we can see from the graph above, machine learning and python are the most popular tags and deep learning, NLP, and neural networks follow close behind. ### Getting Tag views

To determine the number of views for each tag, the following steps are taken:

1. Create an empty dictionary `tag_view_count`.
2. Loop through each row in the `questions` dataframe using pandas [iterrows()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html) method.
3. Within the loop, loop through each tag in the *list of tags* for the current row.
4. Check if the current tag is already in the `tag_view_count` dictionary.
5. If the current tag is already in the dictionary, add the value of the `ViewCount` for that row to the current value for that key.
6. If the current tag is not in the dictionary, add it with a value equal to the `ViewCount` for that row.

````python
# Create an empty dictionary (tag_view_count) to store each tag as a key with its view counts as its value
tag_view_count = {}

for index, row in questions.iterrows():
    for tag in row['Tags']:
        if tag in tag_view_count:
            tag_view_count[tag] += row['ViewCount']
        else:
            tag_view_count[tag] = row['ViewCount']
````

````python
# Create dataframe most_viewed
most_viewed = pd.DataFrame.from_dict(tag_view_count, orient='index', columns=['ViewCount'])
most_viewed
````

````python
most_viewed = most_viewed.sort_values(by='ViewCount', ascending=False).head(20)

ax = most_viewed.sort_values('ViewCount').plot(kind='barh', figsize=(12, 6), color='darkblue')
ax.set_title('Top 20 Most Viewed Tags on DSSE', size=20, weight='bold' )
ax.grid(False)

plt.show()
````

The graph of the 20 most viewed tags on Data Science Stack Exchange reveals that machine-learning, python, and deep-learning are the most popular topics with over 150,000 views each. They are followed by neural-network, nlp, and classification, showing a high level of interest in these areas among the data science community. The most viewed tag was "machine-learning" with 184795 views, followed by "python" with 147303 views and "deep-learning" with 104839 views. The least viewed tags in the top 20 were "cnn" and "regression" with 16053 and 16297 views respectively.

### Tag classification

Let's establish the criteria for identifying deep learning questions. Our definition will consider a question to be related to deep learning if it has one or more of the following tags: `['lstm', 'cnn', 'pytorch', 'tensorflow', 'keras', 'neural-network', 'deep-learning']`.

To arrive at this list, we analyzed all the tags in the `most_used` list and selected the ones related to deep learning. This can also be confirmed through further research and reading the descriptions of these tags on the internet.

We will now write a function that assigns a value of `1` to questions that meet the above criteria, and `0` otherwise. This function will be applied to the dataframe.

````python
def tag_category(tags):
    for tag in tags:
        if tag in ['lstm', 'cnn', 'pytorch', 'tensorflow',
                   'keras', 'neural-network', 'deep-learning']:
            return 1
    return 0
````

````python
all_questions['DeepLearning'] = all_questions['Tags'].apply(tag_category)
all_questions.head()
````

````python
year_group = dl_questions.groupby('Year').agg({"DeepLearning": ['sum', 'size']})
year_group.columns = ['DeepLearningQuestions', 'TotalQuestions']
year_group['DeepLearningRate'] = round(year_group['DeepLearningQuestions'] / year_group['TotalQuestions'], 2)*100

# Reset the index to make 'Year' as column
year_group.reset_index(inplace=True)
year_group
````

````python
fig, ax = plt.subplots(figsize=(15, 8))
bar1 = ax.bar(year_group['Year'], year_group['TotalQuestions'], label='total questions', color='#bcc6cc', alpha=0.7)
bar2 = ax.bar(year_group['Year'], year_group['DeepLearningQuestions'], label='deep learning questions', color='#98afc7')

for rect, i in zip(bar2, year_group['DeepLearningRate']):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2.0, height, f'{i:.0f}%', ha='center', va='bottom')
for rect in bar1:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2.0, height, f'{height:.0f}', ha='center', va='bottom')

ax.set_xticks(year_group['Year'])
ax.set_yticks([])
ax.set_ylim(0, 7500)
ax.set_title('Tracking the Rise of Deep Learning Interest from 2014 to 2022', size=16, weight='bold')
ax.text(2015, 7200, '37% of the total questions asked on DSSE in 2018 were deep learning questions', size=12)
ax.legend()
ax.grid(False)

plt.show()
````
