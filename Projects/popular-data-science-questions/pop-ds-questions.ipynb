Now let's address the issue we discovered in the `Tags` column by cleaning it to meet our needs.

The values in the `Tags` column are strings that look like this:

```
"<machine-learning><regression><linear-regression><regularization>"
```

We will clean the string format in the `Tags` column to make it easier to use string methods. Our aim is to change strings in the format as shown above to something like:

```
"machine-learning,regression,linear-regression,regularization"
```

We can then use the `,` as a separator to convert the string into a list.

````python
# Transform '<' to an empty string and '>' to ',' while ensuring the last tag has no ','
questions['Tags'] = questions.Tags.str.replace('<', '').str.replace('>', ',').str.rstrip(',').str.split(',')
````

## Most Used and Most Viewed

We are now focused on determining the most popular tags. To do this, we'll consider two different proxies for popularity. The first proxy is the number of times a tag was used. The second proxy is the number of times a question with that tag was viewed.

In order to further refine our research, we could take into account other factors such as the score of the question or whether it is a part of someone's favorite questions. However, for now, we'll limit our focus to counts and views as our proxies for popularity.

### Getting Tag count

Here, the goal is to count how often each tag was used. The following steps will be taken:

1. We start by creating an empty dictionary called `tag_count`.
2. Then, we loop through each *list of tags* in the `Tags` column of the `questions` dataframe.
3. For each *list of tags*, we loop through each tag within the list.
4. For each tag, we check if it already exists in the `tag_count` dictionary.
5. If the tag is already in the dictionary, we add 1 to its count.
6. If the tag isn't in the dictionary yet, we add it and start its count at 1.

````python
# Create an empty dictionary (tag_count) to store each tag as a key with its count as its value
tag_count = {}

for tags in questions['Tags']:
    for tag in tags:
        if tag in tag_count:
            tag_count[tag] += 1
        else:
            tag_count[tag] = 1
````

Let's turn `tag_count` into a dataframe (`most_used`) for a better display.

````python
# Create dataframe most_used
most_used = pd.DataFrame.from_dict(tag_count, orient='index', columns=['Count'])
most_used
````

Get the top 20 most used tags in the `most_used` dataframe.

````python
most_used = most_used.sort_values(by='Count', ascending=False).head(20)
most_used
````

Let's visualize the top 20 most used tags.

````python
ax = most_used.sort_values('Count').plot(kind='barh', figsize=(12, 6), color='darkblue')
ax.set_title('Top 20 Most Popular Tags on DSSE', size=20, weight='bold' )
ax.grid(False)

plt.show()
````

As we can see from the graph above, machine learning and python are the most popular tags and deep learning, NLP, and neural networks follow close behind. This highlights the significance of these topics in the field of data science and machine learning. Additionally, sickit-learn, Tensorflow, Keras, and PyTorch are popular machine learning/deep learning libraries that are widely used.

### Getting Tag views

To determine the number of views for each tag, the following steps are taken:

1. Create an empty dictionary `tag_view_count`.
2. Loop through each row in the `questions` dataframe using pandas [iterrows()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html) method.
3. Within the loop, loop through each tag in the *list of tags* for the current row.
4. Check if the current tag is already in the `tag_view_count` dictionary.
5. If the current tag is already in the dictionary, add the value of the `ViewCount` for that row to the current value for that key.
6. If the current tag is not in the dictionary, add it with a value equal to the `ViewCount` for that row.

````python
# Create an empty dictionary (tag_view_count) to store each tag as a key with its view counts as its value
tag_view_count = {}

for index, row in questions.iterrows():
    for tag in row['Tags']:
        if tag in tag_view_count:
            tag_view_count[tag] += row['ViewCount']
        else:
            tag_view_count[tag] = row['ViewCount']
````

````python
# Create dataframe most_viewed
most_viewed = pd.DataFrame.from_dict(tag_view_count, orient='index', columns=['ViewCount'])
most_viewed
````

````python
most_viewed = most_viewed.sort_values(by='ViewCount', ascending=False).head(20)

ax = most_viewed.sort_values('ViewCount').plot(kind='barh', figsize=(12, 6), color='darkblue')
ax.set_title('Top 20 Most Viewed Tags on DSSE', size=20, weight='bold' )
ax.grid(False)

plt.show()
````
